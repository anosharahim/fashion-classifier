```python
import zipfile
import os
import numpy as np
from PIL import Image
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

```python
#extract from zipped files 
with zipfile.ZipFile("female-clothing.zip", 'r') as zip_ref:
    zip_ref.extractall()
with zipfile.ZipFile("male-clothing.zip", 'r') as zip_ref:
    zip_ref.extractall()
    
    
#get image file paths 
m_files = os.listdir('../fashion-classifier/men')
w_files = os.listdir('../fashion-classifier/women')
print('m:',len(m_files),'| w:',len(w_files))

def read_img(file, name):
    '''Takes in file and path_name and returns a list of image arrays
    '''
    images = []
    for image in file: 
        arr = Image.open(name+'/'+image)
        arr = np.array(arr)
        f_arr = arr.flatten()
        images.append(f_arr)
    return images
 
#read images from fils 
m_images = read_img(m_files,'men') #list of image arrays
w_images = read_img(w_files,'women') #list of image arrays
```
